```python
import numpy as np
import matplotlib.pyplot as plt
```

'oblique_datacube.npz' contains 4 data cubes (3 velocity data cubes and 1 position data cube)
- vel20 means velocity profiles at 20 degree impact angle 
- velocity data cubes should all have a dimension of (36,4,2000)
    - There are 36 spacial data points for each velocity data cubes -> 1st array dimension 
    - 4 accelerometer axes: time, x, y, z -> 2nd array dimension 
        - x is radially outward from the impact point (parallel to level surface)
        - y is in verticle direction (anti parallel to the direction of gravity)
        - in theory we don't expect pulses in z direction 
    - 2000 data points -> 3rd array dimension 
    - velocity is in unit of m/s
    - time is in milisecond
    - t=0 denotes the impact time 
- pos means position data cube. 
- the position data cube has a dimension of (36,2)
    - There are 36 spacial data points -> 1st array dimension 
    - 2 positional axes: x,z -> 2nd array dimension
    - x is the horizontal axis and z is the vertical 
    - y = 0 for all time
    - x,z are in the unit of centimeter

```python
loaded_data = np.load('oblique_datacube.npz')
vel20 = loaded_data['vel20']
vel40 = loaded_data['vel40']
vel60 = loaded_data['vel60']
pos   = loaded_data['pos']
print(np.shape(vel20))
print(np.shape(pos))
```

```python
## example: 40 degree impact angle - velocity x component & z component
plt.plot(vel40[0,0,:],vel40[0,1,:],label='x')
plt.plot(vel40[0,0,:],vel40[0,2,:],label='y')
plt.legend()
plt.xlabel('Time (ms)', fontsize = 15)
plt.ylabel('velocity (m/s)', fontsize = 15)
```

```python
## example: accelerometer placement 

plt.plot(pos[:,0],pos[:,1],'o')
plt.xlabel('R (cm)', fontsize = 15)
plt.ylabel('z (cm)', fontsize = 15)
```
